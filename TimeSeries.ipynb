#Stationarity
A time series is stationary if it does not exhibit any long term trends or obvious seasonality

Constant variance and mean through time

The statistical properties of the time series does not change

To make time series stationary:
    differencing= y(t-1) - y(t) -> this will make the mean zero
    to stabilize the variance, apply log transform = natural log(data)
    to stabilize both, stabilize variance first and then mean = log then diff

We do not stabilize mean first, because diff assumes zero variance.
So, this could end up not making the data stationary.


Stationarity test: ADF test (Augmented dickey fuller test)- Hypothesis test where the null hypothesis is the series is non-stationary
also known as unit root test.-> scipy adfuller library.


If the adf test score < any critical score then we say we are (Criticla score percent value)% sure that the time series is stationary. The ADF test simply checks if the regression coefficeient equals zero.

# Box-Cox Transform
We need constant variance.

The normal log transformation assumes the data came from exponential trend.

For most time series models, underlying assumption is the data come from normal distribution.

The box-cox transformation tranforms non-normal distribution to normal distribution.

Box-Cox: lambda = -5 to 5, need to find the best lambda that makes the data more look like normal distribution.

scipy.stats import boxcox -> gives you the best lambda for your data.

Boxcox is better than logarithm transformation since it finds the best base for logarithm.

# Seasonality

Correct order:
- box-cox
- Remove seasonality
- Remove trends (regular diff)

Seasonality in a time series is a seasonal fluctuations in data. Most time series model expects stationary data, hence we usually need to get rid of seasonality. Overall, we need to be aware of a seasonality component in the data and try to use it in our advantage.

## Remove seasonality:
diff between current value and the value of the previous season (hours, weeks, months, years, etc)



# Time Series Decomposition
Trend: overall motion of the series
seasonality: regular seasonal patterns
Residual/remainder: the bit left over after we take into account the trend and seasonality.

Additive model = T + S + R
multiplicative model = T * S * R


Convert a multiplicative model to an additive model by taking the log or the box-cox transform.

## How is decomposition done:
 - Compute the trend, T, using a moving/rolling average.
 - De-trend the series, Y-T for additive and Y/T for multiplicative
 - Compute the seasonal component, S, by taking the average of the de-trended series for season.
 - The residula component, R = Y-T-S for additive models and R = Y/(T*S) for multiplicative components.

 Other techniques: not required that much, STL, X11, SEATS.

# Autocorrelation

In time series models, we make inferences about the past to produce forecast about the future.

Autocorrelation helps us detect certain features in our series to enable us to choose the most optimal forecasting model for our data.


Autocorrelation is the correlation of the data with itself. Instead of measuring the correlation between two random variables, we measure the correlation between a random variable against itself. Autocorrelation is correlation of the time series at two different points in time (lags). We measure time series against lagged version of itself.

The autocorrelation between y_t and y_t is 1 because they are identical.


## Reasons:
- Seasonality identification: if we find the correlation at certain lag multiples is in general higher than others, we say there is a seasonality.
- Trend identify: Correlation for recent lags is higher and slowly decreasing as the lags increase, then there is trend in our data.


## Decode the Autocorrelation plot:
- slow decay across many lags - non-stationary/trend
- sharp cutoff after lag q - MA (Moving Average) behavior
- exponential or dampening decay- AR (Autoregressive) behavior
- spikes at fixed intervals - seasonality
- Mostly zero after lag 0 - white noise


The bar outside the bands -> significant correlation, inside bands -> not significant correlation.

# Partial Autocorrelation

The correlation between the time series at two different lags not considering the effect of any intermediate lags.

PACF reveals direct lag dependencies that are hidden behind overall trends or indirect correlations in the data.

# Evaluating time series forecasting
1. Mean Absolute Error (MAE)
- Pros:
  - Intuitive
  - Error same unit as forecast
- Cons:
  - Doesn't penalise outliers
  - Scale dependent
2. Mean Square Error (MSE)
- Pros:
  - Penalise outliers
- Cons:
  - Less interpretable
  - Scale dependent

3. Root Mean Square Error (RMSE)
- Pros:
  - Penalise outliers
  - Error in forecast unit
  - Best of MAE and MSE
- Cons:
  - Less interpretable
  - Scale dependent

4. Mean Absolute Percentage Error (MAPE)
- Pros:
  - Easy to interpret as percentage
  - Scale independent
- Cons:
  - Infinite error if the actual value is near zero
  - Biased to under-forecast

5. Symmetric Mean Absoulute Percentage Error (SMAPE)
- Pros:
  - No longer favors under forecasting
- Cons:
  - Infinite error if the actual value is near zero
  - Hard to interpret
  - Not actually symmetric

2. Mean Squared Logarithm Error (MSLE)
- Pros:
  - Punishes under-forecasting
- Cons:
  - Divinding by values close to zero
  - Hard to interpret

# Basic Forecasting Methods

Simple models provides a baseline for the overall problem. Otherwise, we will end up wasting too much resources with overly complecated models, when we only need a simple one, in reality.

## Average model
All futurre values (test data) are equal to the mean of all the previous observations (train data).

## Naive model
Set the future forecast to the latest observed value

## Seasonal Naive model
Forecast is equal to the most recent observed value in the same season. Forecast for the next quarter 1 is equal to quarter 1 of the past year.

## Drift model
Extension of the naive forecast. We let the prediction either linearly increase ir decrease through time as a function of time step, h, scaled by the average historical trend.

# Simple exponential Smoothing

This model does not take into accout trends or seasonality. You can transform with differencing and box-cox.

This model gives more weights to the most recent values while exponentially decreasing weights for the values more in the past, thus providing a smoothing effect.

```python
# Import packages
import plotly.graph_objects as go
import pandas as pd
from statsmodels.tsa.holtwinters import SimpleExpSmoothing

# Read in the data
data = pd.read_csv('AirPassengers.csv')
data['Month'] = pd.to_datetime(data['Month'])

data['#Passengers'] = np.log(data['#Passengers'])
data['#Passengers'] = data['#Passengers'].diff(12)
data = data.dropna()

# Split train and test
train = data.iloc[:-int(len(data) * 0.2)]
test = data.iloc[-int(len(data) * 0.2):]


def plot_func(forecast: list[float], title: str) -> None:
    """Function to plot the forecasts."""
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=train['Month'], y=train['#Passengers'], name='Train'))
    fig.add_trace(go.Scatter(x=test['Month'], y=test['#Passengers'], name='Test'))
    fig.add_trace(go.Scatter(x=test['Month'], y=forecast, name='Forecast'))
    fig.update_layout(template="simple_white", font=dict(size=18), title_text=title,
                      width=650, title_x=0.5, height=400, xaxis_title='Date',
                      yaxis_title='Passenger Volume')

    return fig.show()


# Fit model and get forecasts
model = SimpleExpSmoothing(train['#Passengers']).fit(optimized=True)
forecasts = model.forecast(len(test))
print(forecasts)

# Plot the forecasts
plot_func(forecasts, 'Simple Exponential Smoothing')
```
